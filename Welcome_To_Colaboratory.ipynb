## Hızlandırıcılar
Colaboratory, ücretsiz GPU ve TPU (Tensör İşleme Ünitesi) hızlandırıcıları sunar.

*Runtime > Change runtime type* türünü değiştir menüsünü kullanın, bir TPU hızlandırıcı seçin ve ardından **not defterini yeniden çalıştırın.**

```
try:
  tpu = tf.contrib.cluster_resolver.TPUClusterResolver() # TPU detection
  strategy = tf.contrib.tpu.TPUDistributionStrategy(tpu)
except ValueError:
  print("Running on GPU or CPU")
```

Sekiz TPU çekirdeğini gördünüz mü?
